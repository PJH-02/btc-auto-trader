```python
import yfinance as yf

# 비트코인 가격 데이터 다운로드 (5년치 데이터)
btc = yf.download('BTC-USD', period='5y', interval='1d')

# 데이터 확인
print(btc.head())
```

```python
# 멀티인덱스에서 'Ticker' 레벨을 드롭하고, 컬럼명을 설정
btc.columns = ['Close', 'High', 'Low', 'Open', 'Volume']

# 인덱스를 리셋하고 'Date' 컬럼을 추가
btc = btc.reset_index()

# 'Date', 'Close', 'High', 'Low', 'Open', 'Volume' 순서로 컬럼 재정렬
btc = btc[['Date', 'Close', 'High', 'Low', 'Open', 'Volume']]
```
